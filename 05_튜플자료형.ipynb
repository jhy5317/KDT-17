```python
# 튜플은 어떻게 만들까? (tuple)
# 튜플은 수정과 삭제가 안됨, 1개의 요소를 적어도 콤마 필요

t1 = ()
t2 = (1, )
t3 = (1, 2, 3)
t4 = 1, 2, 3 #튜플은 괄호 생략도 가능
t5 = ("a", "b", ("ab", "cd"))

print(type(t1))
print(type(t2))
print(type(t3))
print(type(t4))
print(type(t5))

print(t1)
print(t2)
print(t3)
print(t4)
print(t5)
```

```python
a = 1, # 콤마찍으면 튜플이 됨.

print(type(a)) # 콤마없이 하나의 요소로 하면 튜플이 안됨;
```

```python
# 튜플의 요소값을 지우거나 변경하려고 하면 어떻게 될까?

t1 = 1, 2, "a", "b"

# del t1[0] 안됨.
# 변경하려고 할때
# t1의 첫번째 인덱스인 0을 변경하기

#t1[0] = "c" <- 안됨
```

```python
# 튜플 인덱싱

t1 = 1, 2, "a", "b"

print(t1[2])
print(t1[-1])
```

```python
# 튜플의 슬라이싱

t1 = 1, 2, "a", "b"

print(t1[1:])
print(t1[2:])
```

```python
# 튜플 더하기

t1 = 1, 2, "a", "b"
t2 = 3,4

a = t1+t2

print(a)
print(type(a))
```

```python
# 튜플 곱하기

t2 = 3,4

print(t2*3)
print(t2)
```

```python
# 튜플의 길이

t2 = 3,4

print(len(t2))
```
