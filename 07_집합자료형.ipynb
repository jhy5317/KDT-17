```python
# 집합자료형은 어떻게 만들까?

s1 = set([1, 2, 3])
s2 = {4, 5, 6}

print(s1)
print(type(s1))

print(s2)
print(type(s2))
```

```python
s3 = set("hello")
print(s3) # 각각 문자가 중복없이 랜덤으로 들어감.
print(type(s3))
```

```python
# 집합자료형의 특징
## 1. 중복을 허용하지 않는다.
## 2. 순서가 없다.
```

```python
# 교집합 구하기 -> intersection

s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print(s1)
print(s2)

## intersection() 사용
print(s1.intersection(s2))

## & 사용
print(s1 & s2)
```

```python
# 합집합 구하기 -> union

s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print(s1)
print(s2)

## union()
print(s1.union(s2))

## | 사용
print(s1 | s2)
```

```python
# 차집합 구하기

## difference 사용
print(s1.difference(s2))
print(s2.difference(s1))

## - 사용
print(s1 - s2)
print(s2 - s1)
```

```python
# 집합 자료형에 관련된 함수

## 값 1개 추가하기 -> add
s1 = {1, 2, 3}

print(s1)
s1.add(4)
print(s1)
```

```python
## 값 여러개 추가하기 -> update

s1 = {1, 2, 3}
s1.update([4,5,6])

print(s1)
```

```python
## 특정 값 제거하기 -> remove

s1 = {1, 2, 3}
s1.remove(1)

print(s1)
```

```python
## 모든 값 제거하기 -> clear

s1 = {1, 2, 3}

s1.clear()

print(s1)

# s1 = set()은 빈 집합자료형
# s2 = {}은 딕셔너리가 됨. 빈 집합은 set()로 표현.
```
