```python
# 문자열은 어떻게 만들고 사용할까?

## 1. 큰따옴표로 양쪽 둘러싸기
a = "Hello World"

## 2. 작은따옴표로 양쪽 둘러싸기
b = 'Python is fun'

## 3. 큰따옴표 3개로 둘러싸기
c = """Life is too short, You need python"""

## 4. 작은따옴표 3개로 둘러싸기
d = '''Life is too short, You need python'''

print(a)
print(b)
print(c)
print(d)

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
# 문자열안에 작은 따옴표 포함하기

a = "Python's favorite food is perl"
b = """Python's favorite food is perl"""
c = 'Python\'s favorite food is perl'

print(a)
print(b)
print(c)
```

```python
# 문자열안에 큰따옴표 포함하기

#a = "Python is very easy" he says."
a = '"Python is very easy" he says.'
b = "\"Python is very easy\" he says."

print(a)
print(b)
```

```python
multiline = """Life is too short, 
You need Python"""
multiline2 = "Life is too short, \nYou need Python"

print(multiline)
print(multiline2)

```

```python
# 이스케이프 코드

# \n -> 문자열 안에서 줄 바꿀 때
# \t -> 문자열 안에서 간격을 줄 때
# \\ -> 문자열 안에서 역슬래시(\)를 표현할 때
# \' -> 문자열 안에서 작은 따옴표 표현할 때
# \" -> 문자열 안에서 큰 따옴표 표현할 때
```

```python
# 문자열 연산하기

a = "Python "
b = "is fun"

print(a+b)
```

```python
a = "Python"
print(a*3)
```

```python
# 문자열의 길이 구하기

a = "Life is too short"

print(len(a))
```

```python
# 문자열 인덱싱

a = "Life is too short, You need Python."

print(a[3])
print(a[8])
print(len(a))
print(a[-1]) # 마지막 문자 호출.
print(a[-3]) # 뒤에서 3번째 분자 호출.
```

```python
# 문자열 슬라이싱

a = "Life is too short, You need Python."

print(a[12:17])
print(a[:])
print(a[5:])
```
