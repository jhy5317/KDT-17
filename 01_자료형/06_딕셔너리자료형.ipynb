```python
# 딕셔너리는 어떻게 만들까?
# {key1: value1, value2, value3, ...}

dic = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}

print(dic)
print(type(dic))

# 딕셔너리에 key값으로 올 수 없는 데이터 타입 -> list (변경이 가능하기 때문임.)
```

```python
# 딕셔너리 쌍 추가하기

a = {1 : 'a'}
a[2] = 'b' # 딕셔너리에서의 [2]는 인덱스가 아니라 key이다.

print(a)

a[3] = [1,2,3]
print(a)

# 딕셔너리 요소 삭제하기

print(a)
del a[1]

print(a)
```

```python
# 딕셔너리에 key를 사용해서 value 얻기

grade = {
	'pey': 10,
	'juli': 99
}

print(grade['pey'])
print(grade['juli'])
```

```python
# 딕셔너리 관련 함수

## 1. key 리스트 만들기 -> keys
a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}

a_key = list(a.keys())
print(a_key)

## 2. value 리스트 만들기 -> values
a_value = list(a.values())
print(a_value)
```

```python
# key, value 쌍 얻기 -> items

print(a)
a.items()

print(list(a.items())) # 리스트안의 하나의 요소요소는 튜플로 되어있음.
```

```python
# key, value 쌍 모두 지우기 -> clear

a.clear()
print(a)
```

```python
# key로 value 얻기 -> get

a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}

print(a["name"])
print(a.get("name"))

# print(a["hobby"]) -> 없는 키는 에러
print(a.get("hobby",0)) # get함수를 쓰면 키가 없어도 반환값 설정 가능
existKey = a.get("name", "이 키는 존재하지않습니다.") # 있으면 value를 반환하고 없으면 뒤의 문자열 출력.
print(existKey)
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 -> in

print(a)

print('name' in a)
print('email' in a)
```

```python
# key로 value 꺼내기 -> pop

a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}

result = a.pop("phone")

print(result)
print(a)
```
