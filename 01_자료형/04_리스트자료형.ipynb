```python
# 리스트는 어떻게 만들고 사용할까?

# 리스트명 = [요소1, 요소2, ...]

a = []
b = [1, 2, 3]
c = ["Life", "is", "too", "short"]
d = [1, 2, "Life", "is"]
e = [1, 2, ["Life", "is"]]

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
# 리스트의 인덱싱과 슬라이싱

a = [1, 2, 3]

print(a[0])
print(type(a[1])) # 리스트 안에있어도 각각의속성을 따름.

print(a[0]+a[2]) # 1+3
print(a[-1]) # 똑같이 마지막 값 반환.
```

```python
b = ["1", "2", "3", "a", "b", "c"] #각각 내용물은 문자열임. 숫자x

print(b[-1])

print(b[0]+b[2]+b[-1]) # 문자열이기때문에 이어붙이기 연산을 함.
```

```python
c = [1, 2, ["a", "b", ["Life", "is"]]]

print(len(c))

print(c[-1][-1][-1]) 
```

```python
# 리스트의 슬라이싱

a = [1, 2, 3, 4, 5]

print(a[0:2])
print(a[3:4])
print(a[2:])
print(a[:3])
```

```python
a = [1, 2, 3, ["a", "b", "c"], 4, 5]

# ['b', 'c'] 호출
print(a[3][1:])
```

```python
# 리스트의 연산

## 더하기
a = [1, 2, 3]
b = [4, 5, 6]

print(a+b)

## 곱하기 <- 문자열 취급.

c = [1, 2, 3]
print(3 * c)
```

```python
# 리스트의 수정과 삭제

a = [1, 2, 3]

## [1, 2, 3] -> [1, 2, 4]
a[2] = 4
print(a)

## del 명령어로 리스트의 요소 삭제
b = [1, 2, 3]
del b[1]

print(b)

## 슬라이싱을 이용한 삭제

c = [1, 2, 3, 4, 5]
del c[2:]

print(c)
```

```python
# 리스트 관련 함수

## 리스트 요소 추가하기 -> append
a = [1, 2, 3]
### a에 4라는 값 추가하기
a.append(4) #맨뒤에 추가됨.

print(a)

### 리스트 추가하기
a.append([5, 6])
print(a)
```

```python
# 리스트 정렬 -> sort

a = [1, 4, 2, 3]

a.sort()
print(a)

a.sort(reverse=True)
print(a)
```

```python
a = [1, 4, 2, 3]
print(a)

sorted(a,reverse=True)
```

```python
# 리스트 뒤집기 -> reverse

a = ["a","c","b"]

a.reverse()

print(a)
```

```python
# 인덱스 반환 -> index

a = ["a","c","b"]

print(a.index("c"))
```

```python
# 리스트에 요소 삽입 -> insert

a = [1, 2, 3]

a.insert(0,4)

print(a)
```

```python
# 리스트에 요소 제거 -> remove

a = [1, 2, 3, 1, 2, 3]

a.remove(3)
print(a)
a.remove(3)
print(a)
```

```python
# 리스트의 요소 꺼내기 -> pop

a = [1, 2, 3]
result = a.pop()

print(a) # 리스트에서 빠져나감.
print(result) # 리스트가아닌 속성으로 반환
```

```python
# 리스트에 포함된 요소 x의 개수 세기 -> count

a = [1, 2, 3, 1]

print(a.count(3))
```

```python
# 리스트 확장 -> extend

a = [1, 2, 3]

a.extend([4,5])

print(a)
```
