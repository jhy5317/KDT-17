```python
print(1 == 1)
print(2 > 1)
print(2 < 1)
print(2 >= 2)
print(2 <= 2)
print(2 != 2)
```

```python
# if 조건문(참과 거짓을 판별함):
#	수행할 문장 ~

print(bool('python')) # 문자열이 들어가면 True
print(bool("")) # 비어있는 문자열이 들어가면 False
print(bool(" ")) # 공백도 True

print(bool(10)) # 숫자는 0빼고 전부 True
print(bool(0))

print(bool([1, 2, 3]))
print(bool([])) # 비어있는 리스트는 False

print(bool({"name": "Pey"}))
print(bool({})) # 비어있는 딕셔너리도 False
```

```python
# 논리 연산자

## and 연산자

print(True and True)
print(True and False)
print(False and True)
print(False and False)
print(True and True and False)
```

```python
## or 연산자

print(True or True)
print(True or False)
print(False or True)
print(False or False)
print(True or True or False)
```

```python
## not 연산자

print(not True)
print(not False)
```

```python
x = 20
print(x >= 10 and x < 20)

y = 3
print(y <= 7 or y > 65)
```
