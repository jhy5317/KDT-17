```python
# 문자열 포매팅
## 1. 숫자 바로 대입

print('I eat %d apples' % 3)

## 2. 문자열 바로 대입

print('I eat %s apples' % 'five')
```

```python
## 3. 숫자값을 나타내는 변수로 대입

number = 3
print("I eat %d apples." % number)
```

```python
# 2개 이상의 값 넣기

number = 10
day = 'three'

print("I eat %d apples, so I was sick for %s days" %(number,day))
```

```python
# format 함수를 사용한 포매팅

print("I eat {0} apples".format(3))
print("I eat {0} apples".format('five'))
```

```python
# format 함수를 이용해서 2개이상의 값 넣기

number = 10
day = 'three'
print("I eat {0} apples, so I was sick for {1} days".format(number, day))
```

```python
# f문자열 포매팅(잘 알아두기 거의 이것만 씀)

name = "홍길동"
age = 10

print(f"나의 이름은 {name}입니다. 나이는 {age}입니다.")
```

```python
# 문자열 관련 함수들

a = "hobby"

## 문자의 개수를 세는 함수 count
print(a.count("b"))

## 문자의 위치를 알려주는 함수 find

b = "Python is the best choice"
print(b.find("b"))
print(b.find("k")) # 없으면 -1을 반환함.

## 문자의 위치를 알려주는 함수2 index <- 얘는 없으면 에러.
c = "Life is too short"
print(c.index("t")) # 여러개 있으면 처음 값 반환.
```

```python
# 문자열 삽입하는 함수 join

print(",".join("abcd"))
```

```python
# 소문자를 대문자로 바꾸는 함수 upper

a = "hi"
print(a.upper())

# 대문자를 소문자로 바꾸는 함수 lower

b = "HI"
print(b.lower())
```

```python
# 공백 지우기 strip

a = "   hi   "

print(a)
print(a.strip())
```

```python
# 왼쪽/오른쪽 공백 지우기 lstrip, rstrip

a = "   hi   "

print(a.lstrip())
print(a.rstrip())
```

```python
# 문자열 바꾸기 replace

a = "Life is too short"

print(a.replace("Life","Your leg"))
```

```python
# 문자열 나누기 split -> 나누었을 때 결과값이 중요.

a = "Life is too short"

print(a.split()) # 인수가 비어있으면 공백을 기준으로 나눔.
```
