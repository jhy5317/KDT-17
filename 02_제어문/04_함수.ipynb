```python
# 함수의 구조

# def 함수이름():
# 	수행할 문장1
# 	수행할 문장2
# 	...

def add(a,b):
	return a+b

print(add(5,15))
```

```python
# 1. 일반적인 함수
# 매개변수 o, 반환값 o

# def 함수이름(매개변수):
# 	수행할 문장
# 	...

# 	return 반환값

def add(a,b):
	result = a + b
	return result

a = add(3,4)
print(a)
```

```python
# 2. 입력값이 없는 함수
# 매개변수 x, 반환값 o

def say():
	return 'hi'

a = say()
print(a)
```

```python
# 3. 반환값이 없는 함수
# 매개변수 o, 반환값 x

def add(a, b):
	print(f"{a}, {b}의 합은 {a+b}")

a = add(4,9) # -> add의 함수자체의 print가 실행된 결과
print(a) # -> a의 값은 함수의 return이 없기때문에 None

# 기본적으로 return을 안적어도 return None이 생략되어있음.
```

```python
# 4. 입력값도 없고, 반환값도 없는 함수
# 매개변수 x, 반환값 x

def say():
	print("Hi")

a = say()
print(a)	
```

```python
# 매개변수를 지정하여 호출하기
# 매개변수가 많을때, 특정 부분만 넣고 함수를 호출하기위해 사용함.

def sub(a, b):
	result = a - b
	return result


c = sub(a = 7, b = 3)
print(c)

d = sub(b = 29, a = 28)
print(d)
```

```python
# 입력값이 몇개가 될 지 모를때는 어떻게 해야할까?

# def 함수이름(*매개변수):
# 	수행할 문장...

def addMany(*args):
	result = 0
	for i in args:
		result += i

	return result

t = addMany(1, 2, 3, 4, 5)
print(t)

print("Hi\n", "Hello\t", "안녕?")
```
