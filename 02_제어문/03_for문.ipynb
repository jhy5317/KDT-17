```python
# for문의 기본 구조

# for 변수 in 리스트 또는 튜플 또는 문자열:
# 	수행할 문장1
# 	수행할 문장2
```

```python
test_list = ["one", "two", "three"]

for tl in test_list:
	print(tl)
```

```python
a = [(1, 2), (3, 4), (5, 6)] # 딕셔너리를 items 메서드를 이용해 값을 받아오면 리스트안의 튜플형태로 나옴.

for (k, j) in a:
	print(j)
```

```python
marks = [90, 25, 67, 45, 80]

number = 0

for mark in marks:
	number += 1
	if mark >= 60:
		print(f"{number}번 학생은 합격입니다.")
	else:
		print(f"{number}번 학생은 불합격입니다.")
```

```python
# for문과 continue

marks = [90, 25, 67, 45, 80]

number = 0

for mark in marks:
	number += 1
	if mark < 60:
		continue
	
	print(f"{number}번 학생은 합격입니다.")
```

```python
# for문과 함께 자주 사용하는 range 함수

a = range(10)

print(list(a))

b = range(1, 10)

print(list(b))

c = range(1, 10, 2)
print(list(c))

d = range(10, 0, -1)
print(list(d))
```

```python
# for와 range를 이용한 구구단

for i in range(2, 4):
	for j in range (1, 10):
		print(f"{i}x{j} = {i*j}")
```

```python
# 리스트 컴프리헨션 (list comprehension) 사용하기

# 방법 1
a = [1, 2, 3, 4]
result = []

for num in a:
	result.append(num*3)

print(result)

# 방법 2
a = [1, 2 ,3 ,4]

result = [num * 3 for num in a] # 방법 1과 같은 효과를 내는 코딩
print(result)
```

```python
# for문과 break

for i in range(10):
	if i == 5:
		break
	else:
		print(i)
```

```python
# enumerate() 함수

fruits = ['apple', 'banana', 'orange']

for fruit in fruits:
	print(f'과일: {fruit}')

for (i, fruit) in enumerate(fruits): # 인덱스를 줌.(튜플형태)
	print(f'{i+1}번째 과일: {fruit}')
```

```python
# zip함수로 여러 리스트 순회하기

names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]

for item in zip(names, scores):
	print(item)
	print(f"{item[0]}의 점수: {item[1]}점")

for (name, score) in zip(names, scores):
	print(f"{name}의 점수: {score}점")
```
