```python
# for문의 기본 구조

# for 변수 in 리스트 또는 튜플 또는 문자열:
# 	수행할 문장1
# 	수행할 문장2
```

```python
test_list = ["one", "two", "three"]

for tl in test_list:
	print(tl)
```

```python
a = [(1, 2), (3, 4), (5, 6)] # 딕셔너리를 items 메서드를 이용해 값을 받아오면 리스트안의 튜플형태로 나옴.

for (k, j) in a:
	print(j)
```

```python
marks = [90, 25, 67, 45, 80]

number = 0

for mark in marks:
	number += 1
	if mark >= 60:
		print(f"{number}번 학생은 합격입니다.")
	else:
		print(f"{number}번 학생은 불합격입니다.")
```

```python
# for문과 continue

marks = [90, 25, 67, 45, 80]

number = 0

for mark in marks:
	number += 1
	if mark < 60:
		continue
	
	print(f"{number}번 학생은 합격입니다.")
```

```python
# for문과 함께 자주 사용하는 range 함수

a = range(10)

print(list(a))

b = range(1, 10)

print(list(b))

c = range(1, 10, 2)
print(list(c))

d = range(10, 0, -1)
print(list(d))
```

```python
# for와 range를 이용한 구구단

for i in range(2, 4):
	for j in range (1, 10):
		print(f"{i}x{j} = {i*j}")
```
