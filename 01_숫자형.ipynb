```python
# 숫자형은 어떻게 만들고 사용할까?

a = 1234
b = -1234
c = 0

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
type(c)
```

```python
# 실수형

a = 1.2
b = -3.14

print(a)
print(b)

print(type(a))
print(type(b))
```

```python
# 8진수와 16진수

a = 0o177
print(a)

1 * (8**2) + 7 * (8**1) + 7 * (8**0)
```

```python
a = 0x8ff

print(a)

8*(16**2) + 15*(16**1) + 15*(16**0)

```

```python
#사칙 연산

a = 3
b = 4

print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
# x의 y제곱을 나타내는 연산자 **

a = 3
b = 4

print(a**b)
```

```python
# 나눗셈 후의 나머지를 나타내는 연산자 %

a = 7
b = 3

print(a % b)
```

```python
# 나눗셈 후의 몫을 반환하는 연산자 //

print(7//3)
```

```python
# 복합 연산자

a = 3

#a = a+1
a += 1

print(a)

```
